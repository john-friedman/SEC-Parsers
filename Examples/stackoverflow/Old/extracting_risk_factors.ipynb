```python
#https://stackoverflow.com/questions/57308830/extraction-of-text-using-beautiful-soup-and-regular-expressions-in-10-k-edgar-fi

from sec_parsers import *

# download html
set_headers("John Doe","johndoe@2mail.com")
html = download_sec_filing('https://www.sec.gov/Archives/edgar/data/1318605/000162828024002390/tsla-20231231.htm')

# parse html
parsed_html = parse_10k(html)

# convert to xml tree
xml = construct_xml_tree(parsed_html)
```

```python
# later on sec-parsers will support selection of nodes by name, for now we can use the following code to get the risk factor section
risk_factor = [node for node in xml.iter() if (('1a' in node.attrib.get('desc','').lower()) & (node.tag=='item'))][0]

# get text
risk_factor_text = get_node_text(risk_factor)
print(risk_factor_text)
```
